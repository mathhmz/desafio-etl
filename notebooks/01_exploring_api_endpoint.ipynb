# Exploracão inicial dos dados.

```python
import pandas as pd
```

```python
#Normalizando o JSON para leitura
df = pd.json_normalize(data['resultado']['listaItem'])
```

```python
df.head(10)
```

```python
df.info(verbose=True)
```

```python
#Definindo quais colunas utilizar e renomeando
columns = {"numero":"number",
          "autor":"author",
          "ano":"year",
          "dataPublicacao":"presentationDate",
          "ementa":"ementa",
          "regime":"regime",
          "situacao":"situation",
          "siglaTipoProjeto":"propositionType",
          "listaHistoricoTramitacoes" : "Tramitacoes" 
}
```

```python
df = df[list(columns.keys())].rename(columns=columns)
df
```

```python
df.info(verbose=True)
```

```python
#Separando o DataFrame de proposições
df_proposicoes = df[['number', 'author', 'year', 'presentationDate', 'ementa', 'regime', 'situation', 'propositionType']].copy()
df_proposicoes['city'] = 'Belo Horizonte'
df_proposicoes['state'] = 'Minas Gerais'
df_proposicoes
```

```python
#Criando o DataFrame separado de tramitações
tram_list = []

for index, row in df.iterrows():
    for tramitacao in row['Tramitacoes']:
        tram_list.append({
            'number': row['number'],
            'createdAt': tramitacao['data'],
            'description': tramitacao['historico'],
            'local': tramitacao['local']
        })

df_tramitacoes = pd.DataFrame(tram_list)
df_tramitacoes
```

# Iniciando a limpeza e as transformações dos dados.

```python
df_proposicoes
```

```python
df_tramitacoes
```

```python
# Convertendo as datas para o formato ISO-8601
df_proposicoes['presentationDate'] = pd.to_datetime(df_proposicoes['presentationDate']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
df_tramitacoes['createdAt'] = pd.to_datetime(df_tramitacoes['createdAt']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
```

```python
# Garantindo que os números das proposições são strings
df_proposicoes['number'] = df_proposicoes['number'].astype(str)
df_tramitacoes['number'] = df_tramitacoes['number'].astype(str)
```

```python
df_tramitacoes.info(verbose=True)
```

```python
#Preenchendo os valores nulos, pois a ementa não é presente em todas as proposições.
#Como em outros campos desses dados, preencho os valores nulos com uma string vazia;
df_proposicoes['ementa'] = df_proposicoes['ementa'].fillna(value="")
```

```python
df_proposicoes.info(verbose=True)
```

```python
#Criando uma função que remove os espacos em branco e busca eliminar qualquer outro caracter especial ou regex que possa causar conflito;
import re
def clean_data(df):
    for col in df.select_dtypes(include=['object']).columns:
        # Remover espaços em branco do início e fim
        df[col] = df[col].str.strip()
        # Remover múltiplos espaços em branco
        df[col] = df[col].str.replace(r'\s+', ' ', regex=True)
        # Remover qualquer outra expressão regular indesejada
        df[col] = df[col].str.replace(r'[^\w\s,.-]', '', regex=True)
    return df

# Aplicar a limpeza nos DataFrames
df_proposicoes_clean = clean_data(df_proposicoes)
df_tramitacoes_clean = clean_data(df_tramitacoes)
```

```python
df_proposicoes = clean_data(df_proposicoes)
```

```python
df_tramitacoes = clean_data(df_tramitacoes)
```
